```python
from openai import OpenAI

client = OpenAI(
    api_key = open(".openai", "r").read()
def geneerateLLMResponse(prompt):
    completion = client.chat.completions.create(
        extra_headers={},
        extra_body={},
        temperature=0,
        model="gpt-4o-mini",
        messages=[
            {
                "role": "developer",
                "content": systemPromptSingleEvaluation
            },
            {
                "role": "user",
                "content": prompt
            }
        ]
    )
    print(completion)
    return completion.choices[0].message.content
```

```python
questionDict = {}
for filename in os.listdir(solutionsBaseFilepath):
    timeStart = time.time()
    try :
        methodDict = {}
        solutionFilepath = solutionsBaseFilepath + str(filename) + "/p1.java"
        studentSampleSolution = open(solutionFilepath).read()
        for method in rubric:
            if type(rubric[method]) == dict:
                methodDict[method] = {}
                for point in rubric[method]:
                    prompt = f"""
                    Question Template: {question}
                    Student Solution: {studentSampleSolution}
                    Method in Question: {method}
                    Rubric Point: {point}"""
                    
                    response = str(geneerateLLMResponse(prompt))
                    if response == "YES":
                        methodDict[method][point] = rubric[method][point]
                    else:
                        methodDict[method][point] = 0
            else:
                point = ["There are no subparts to this method. Evaluate the method as a whole."]
                prompt = f"""
                Question Template: {question}
                Student Solution: {studentSampleSolution}
                Method in Question: {method}
                Rubric Point: {point}"""
                response = str(geneerateLLMResponse(prompt))
                if response == "YES":
                    methodDict[method] = rubric[method]
                else:
                    methodDict[method] = 0
            questionDict[filename] = methodDict
            timeEnd = time.time()
        print(f"Time taken to evaluate {filename}: {timeEnd - timeStart} seconds")
    except FileNotFoundError:
        print(f"Skipping {filename} due to an error")
        continue


with open("./results/jrubric-ptwise_oops_evaluation.json", "w") as f:
    json.dump(questionDict, f, indent=4)
    
```

```python
df2 = pd.read_csv("./files/oops-graded-lvl2.csv")
givenMarks = []
logicalMarks = []
for person in questionDict:
    marksDict = questionDict[person]
    finalMarks = 0
    def getMarks(evaluation: dict):
        global finalMarks
        for key in evaluation:
            if type(evaluation[key]) == dict:
                getMarks(evaluation[key])
            elif type(evaluation[key]) == int:
                finalMarks += evaluation[key]
    getMarks(marksDict)
    print(finalMarks)
    logicalMarks.append(finalMarks)
    graded = df2[df2["folder"] == int(person)]["score-final"].values[0]
    givenMarks.append(int(graded))
```

```python
import scipy.stats, sklearn.metrics
```

```python
spearmanr = scipy.stats.spearmanr(givenMarks, logicalMarks)
kendalltau = scipy.stats.kendalltau(givenMarks, logicalMarks)
kappaScore = sklearn.metrics.cohen_kappa_score(givenMarks, logicalMarks)
print(f"The Spearman R coefficient is {float(spearmanr.statistic)} with p-value {float(spearmanr.pvalue)}")
print(f"The Kendall Tau coefficient is {float(kendalltau.statistic)} with p-value {float(kendalltau.pvalue)}")
print(f"The Cohen Kappa Score is {float(kappaScore)}")
```
